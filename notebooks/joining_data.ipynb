```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
yelp_df = pd.read_csv('yelp_df.csv')
fsqr_df = pd.read_csv('fsqr_df.csv')
ctybikes_df = pd.read_csv('bike_station.csv')
```

```python
#First merging fsqr_df and yelp_df based on category
merged_df = pd.merge(fsqr_df, yelp_df, on=['Category'], how='outer')

merged_df
```

```python
import numpy as np
merged_df['Name'] = merged_df['Name_x'].fillna('') + ' ' + merged_df['Name_y'].fillna('')
merged_df['Latitude'] = merged_df['Latitude_x'].fillna('').astype(str) + ' ' + merged_df['Latitude_y'].fillna('').astype(str)
merged_df['Longitude'] = merged_df['Longitude_x'].fillna('').astype(str) + ' ' + merged_df['Longitude_y'].fillna('').astype(str)
merged_df['Location'] = merged_df['Location_x'].fillna('') + ' ' + merged_df['Location_y'].fillna('')
merged_df['Distance'] = np.where(merged_df['Distance_x'].notnull(), merged_df['Distance_x'], merged_df['Distance_y'])
```

```python
merged_df.keys().unique()
```

```python
new_df = merged_df.drop(['Name_x','Distance_x', 'Location_x', 'Name_y', 'Distance_y','Latitude_x','Longitude_x',
       'Latitude_y','Longitude_y','Location_y', 'location.po_box','review_count'],axis=1)
```

```python
# Replace null values with 0
new_df['Rating'].fillna(-1, inplace=True)  
```

```python
new_df
```

```python
new_df.to_csv("merged_yelp_foursquare_cleaned.csv",index=False)
```

Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
#merging the new_df with ctybikes_df based on Latitude, Longitude
ctybikes_df['Latitude'] = ctybikes_df['Latitude'].astype(float)
ctybikes_df['Longitude'] = ctybikes_df['Longitude'].astype(float)
new_df['Latitude'] = new_df['Latitude'].astype(float)
new_df['Longitude'] = new_df['Longitude'].astype(float)

final_merge = pd.merge(ctybikes_df.round(2), new_df.round(2), on=['Latitude', 'Longitude'], how='inner')
```

```python
final_merge
```

```python
final_merge.drop('lat_long', axis=1, inplace=True)
```

```python
final_merge.rename(columns={'total_bikes': 'Total_bikes'}, inplace=True)
final_merge.rename(columns={'name': 'Bike_station'}, inplace=True)
final_merge.rename(columns={'Name': 'Restaurant_name'}, inplace=True)
```

```python
final_merge
```

```python
final_merge.to_csv('final_merge.csv', index=False)
print("DataFrame saved as CSV successfully!")
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python
#exploring the relation ship between total bikes and distance
sns.regplot(x='Total_bikes', y='Distance', data=final_merge)
plt.show()
```

```python
#exploring the relationships with more variables like rating and distance
sns.pairplot(final_merge[['Total_bikes', 'Rating','Distance']])
```

```python
#assessing the magnitude of the correlation between the different variables
sns.heatmap(final_merge.corr(), annot=True, cmap='coolwarm')
```

Rating and distance have a strong correlation

```python
#analysing the relationship between categy and total number of bikes, 
#and at the same time looking at the mean total number of bikes within each category
sns.barplot(y='Category', x='Total_bikes', data=final_merge, estimator=lambda x: sum(x) / len(x))
```

From the above graph we can see that more bikes are available near Categories 1.Belgian, 2.Seafood, Seafood Markets, 3. Canadian (New), Diners, Bars and 4.Canadian (New). However these categories have longer error bars which indicate high variability in the mean number. On the other handthe category with a very little uncertainty in the mean number calculated is nFloer Store , followed by Japanese, Sushi Bars. The latter 2 categories have also a reasonable amount available near them. 

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3
import pandas as pd
from sqlite3 import Error

# SQLite database
data = sqlite3.connect('mydatabase.db')

# Create the 'bike_station' table
ctybikes_df.to_sql('ctybikes_df', data, if_exists='replace', index=False)

# Create the 'yelp_df' table
yelp_df.to_sql('yelp_df', data, if_exists='replace', index=False)

# Create the 'fsqr_df' table
fsqr_df.to_sql('fsqr_df', data, if_exists='replace', index=False)

# Create the 'final_merge' table
final_merge.to_sql('final_merge', data, if_exists='replace', index=False)

# Close for the database connection
data.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('mydatabase.db')

# Create a cursor object
cursor = conn.cursor()

# Execute your SQL queries
cursor.execute("SELECT * FROM ctybikes_df")
result = cursor.fetchall()
#Execute 2
cursor.execute("SELECT * FROM final_merge")
result = cursor.fetchall()

# Process the query result as needed
for row in result:
    print(row)

# Close the cursor and the database connection
cursor.close()
conn.close()
```
